```python
from robotpy_apriltag import AprilTagDetection, AprilTagDetector, AprilTagPoseEstimator
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img0 = cv2.imread('tag42_36h11.png', 0)
img90 = cv2.rotate(img0, cv2.ROTATE_90_COUNTERCLOCKWISE)
img180 = cv2.rotate(img0, cv2.ROTATE_180)
img270 = cv2.rotate(img0, cv2.ROTATE_90_CLOCKWISE)

at_detector = AprilTagDetector()
at_detector.addFamily("tag36h11")

fig,ax = plt.subplots(1,4, figsize=(20,5))

ax[0].imshow(img0)
item = at_detector.detect(img0)[0]
corners = item.getCorners((0,0,0,0,0,0,0,0))
pairs = np.reshape(corners, [4, 2])
ax[0].plot(pairs[0,0], pairs[0,1], 'o', markersize=40)
ax[0].plot(pairs[3,0], pairs[3,1], 'o', markersize=40)

ax[1].imshow(img90)
item = at_detector.detect(img90)[0]
corners = item.getCorners((0,0,0,0,0,0,0,0))
pairs = np.reshape(corners, [4, 2])
ax[1].plot(pairs[0,0], pairs[0,1], 'o', markersize=40)
ax[1].plot(pairs[3,0], pairs[3,1], 'o', markersize=40)

ax[2].imshow(img180)
item = at_detector.detect(img180)[0]
corners = item.getCorners((0,0,0,0,0,0,0,0))
pairs = np.reshape(corners, [4, 2])
ax[2].plot(pairs[0,0], pairs[0,1], 'o', markersize=40)
ax[2].plot(pairs[3,0], pairs[3,1], 'o', markersize=40)

ax[3].imshow(img270)
item = at_detector.detect(img270)[0]
corners = item.getCorners((0,0,0,0,0,0,0,0))
pairs = np.reshape(corners, [4, 2])
ax[3].plot(pairs[0,0], pairs[0,1], 'o', markersize=40)
ax[3].plot(pairs[3,0], pairs[3,1], 'o', markersize=40)

plt.show()
```
